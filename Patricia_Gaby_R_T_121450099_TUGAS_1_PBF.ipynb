4. Apa yang dimaksud dengan pure functional programming language?

Selain itu, pure functional programming tidak memberikan side effect sehingga tidak bergantung pada variabel global. 6. Membuat program Simple Password encryption, dengan ketentuan
  1. Input password bertipe string dengan batasan panjang password 100 karakter
  2. Setiap karakter dari string tersebut
diubah menjadi ASCII value
  3. Kemudian setiap ASCII value tersebut dikembangkan menjadi 3 value dengan dilakukan operasi berikut:
    1. Value pertama dibagi dengan integer 26 kemudian dibulatkan dan dijumlahkan dengan 80 kemudian diubah kembali menjadi karakter
    2. Value kedua dihitung sisa bagi dengan 26 kemudian dijumlahkan dengan 80 kemudian diubah kembali menjadi karakter
    3. Value ketiga adalah karakter + jika value pertama lebih besar dari value kedua, atau value ketiga adalah karakter - jika value pertama tidak lebih besar dari value kedua
    4. Deretkan value pertama, kedua dan ketiga untuk setiap karakter dalam password tersebut sehingga didapatkan password dalam bentuk baru

```python
def password_encrypt(pw):
  if len(pw) <= 100 and len(pw)!=0:
    ascii = [str(ord(s)) for  s in pw]
    enkripsi_pw = ''.join(chr((int(e)//26)+80)+chr((int(e)%26)+80)+('+' if (int(e)//26)+80 > (int(e)%26)+80 else '-') for  e in ascii)
    print("Enkripsi Password: ", enkripsi_pw)
  else:
    print("Password haruslah sepanjang 1-100 karakter")

def password_decrypt(pw):
  split_password = [(pw[i:i+3]) for i in range(0, len(pw), 3)]
  dekripsi_pw = ''.join(chr(((ord(list(f)[0])-80)*26)+(ord(list(f)[1])-80)) for f in split_password)
  print("Password Asli: ", dekripsi_pw)
```

```python
# Password Encryption
pw = input("Masukkan Password Asli: ")
password_encrypt(pw)
```

```python
# Password Decryption
enkripsi_password = input("Masukkan Enkripsi Password: ")
password_decrypt(enkripsi_password)
```
